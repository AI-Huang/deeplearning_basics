```python
import numpy as np
import matplotlib.pyplot as plt
num_samples = 30000
num_pos = 3000
num_neg = 27000
num_classes = 2
```

```python
pos_weight = num_samples / (num_classes * num_pos)
neg_weight = num_samples / (num_classes * num_neg)
print(pos_weight, neg_weight)
```

```python
def get_class_weights(num_samples=30000, num_pos=3000):
    pos_weight = num_samples / (num_classes * num_pos)
    num_neg = num_samples - num_pos
    neg_weight = num_samples / (num_classes * num_neg)
    return (pos_weight, neg_weight)
```

```python
# test_space = np.arange(1, 30000)
test_space = np.arange(3000, 27000)
weights = np.zeros((len(test_space), 2))
for i, num_pos in enumerate(test_space):
    weights[i, 0], weights[i, 1] = get_class_weights(num_pos=num_pos)
```

```python
print(test_space[2999])
print(weights[2999])
```

```python
plt.plot(test_space, weights[:, 0], label="pos weight")
plt.plot(test_space, weights[:, 1], label="neg weight")
plt.plot(test_space, weights[:, 0]/weights[:, 1], label="pos/neg weight ratio")
plt.legend()
plt.grid()
```
