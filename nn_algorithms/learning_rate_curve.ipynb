```python
# lr = 0.05 * (1+math.cos(float(epoch)/epochs*math.pi))
```

```python
epochs = 400
epoch = np.arange(0, epochs)
```

# Cosine Learning Schedule

```python
base_lr=0.1
lr = base_lr/2 * (1+np.cos(epoch/epochs*np.pi))
```

```python
# Half a period of the cosine function
plt.plot(epoch, lr)
```
