```python
from tensorflow.keras.losses import BinaryCrossentropy
loss = BinaryCrossentropy()
```

```python
y_true = [[0., 1.], [1., 0.]]
y_pred = [[0.6, 0.4], [0.4, 0.6]]
# Using 'auto'/'sum_over_batch_size' reduction type.
bce = tf.keras.losses.BinaryCrossentropy()
cce = tf.keras.losses.CategoricalCrossentropy()
bce(y_true, y_pred).numpy()
```

```python
y_true = np.asarray([0., 1.])
y_pred = np.asarray([0.6, 0.4])
# Using 'auto'/'sum_over_batch_size' reduction type.
bce = tf.keras.losses.BinaryCrossentropy()
cce = tf.keras.losses.CategoricalCrossentropy()
print(f"bce: {bce(y_true, y_pred).numpy()}")
print(f"cce: {cce(y_true, y_pred).numpy()}")
```

```python
## Prepare test space
p = np.linspace(0., 1., 1000) # # class 0 label
num_samples = p.shape[0]
y_true = np.asarray([0., 1.]) # class 0 label, class 1 label
bce_loss = np.zeros(p.shape[0])
cce_loss = np.zeros(p.shape[0])
for i, _p in enumerate(p):
    y_pred = np.asarray([_p, 1-_p])
for i, _p in enumerate(p):
    y_pred = np.asarray([_p, 1-_p])
    bce_loss[i] = bce(y_true, y_pred).numpy()
    cce_loss[i] = cce(y_true, y_pred).numpy()
```

```python
p = np.linspace(0., 1., 1000) # # class 0 label
y_true = np.asarray([0., 1.]) # class 0 label, class 1 label
bce_loss = np.zeros(p.shape[0])
cce_loss = np.zeros(p.shape[0])
for i, _p in enumerate(p):
    logits = np.asarray([_p, 1-_p])
    bce_loss[i] = bce(y_true, logits).numpy()
    cce_loss[i] = cce(y_true, logits).numpy()
# cce_loss2 = 2 * cce_loss
```

```python
plt.figure(figsize=(6, 6))
plt.plot(p, bce_loss, label="BCE loss")
plt.plot(p, cce_loss, label="CCE loss")
# plt.plot(p, cce_loss2, label="CCE loss x 2")
plt.xlabel("p")
plt.ylabel("loss")
plt.xlim(0, 1)
plt.ylim(0, 16)
plt.legend()
plt.grid()
```
